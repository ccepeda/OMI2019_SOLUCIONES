```python
bool esTablero(int x1, int y1, int x2, int y2){
    for (int j = y1; j <= y2; j++){
        for (int i = x1; i <= x2; i++){
            if ((i > x1 && color[i][j] == color[i-1][j]) || // si es del mismo color que el cuadro a su izquierda
                (j > y1 && color[i][j] == color[i][j-1]) || // si es del mismo color que el cuadro de arriba
                (i > x1 && j > y1 && color[i][j] != color[i-1][j-1])){ // si es de diferente color del cuadro en diagonal
                return false;
            }
        }
    }
    return true;
}
```
